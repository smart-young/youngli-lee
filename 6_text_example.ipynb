```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/petitions',sep='|', encoding='utf-8')
```

```python
df.head(1)
```

```python
df['content'] = df['content'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") ## 한글이 아니면 ""로 변환
df['title'] = df['title'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
df.head(1)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(how='any')
df.isnull().sum()
```

```python
#stopword 미리 로딩
filename='data/stopwords'
with open(filename, encoding='utf-8') as f:
    content = f.readlines()
stop = [x.strip() for x in content] 
```

```python
filename='data/stopwords'
with open(filename, encoding='utf-8') as f:
    content = f.read().split("\n")
stop = [x.strip() for x in content] 
```

```python
filename='data/stopwords'
with open(filename, encoding='utf-8') as f:
    content = f.read().split()
```

```python
stop
```

```python
TfidfVectorizer?
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(analyzer = 'word', 
                             #tokenizer = lambda x:x.split(),
                             preprocessor = None, 
                             stop_words = stop, 
                             min_df = 2, # 토큰이 나타날 최소 문서 개수
                             #ngram_range=(1, 2),
                             max_features = 1000
                            )

#%time 
train_data_feature = vectorizer.fit_transform(df['content'])
train_vector = pd.DataFrame(train_data_feature.toarray(), columns=vectorizer.get_feature_names())
print(train_data_feature.shape)
print(train_vector.shape)
```

```python
dic = []

for title  in df['title']:
    dic.append(title.split())
```

```python
dic
```

```python
#포스태깅된 컨텐츠를 100차원의 벡터로 바꿔라. 
#주변 단어(window)는 앞뒤로 두개까지 보되, 
#코퍼스 내 출현 빈도가 2번 미만인 단어는 분석에서 제외해라. 
#CPU는 쿼드코어를 쓰고 100번 반복 학습해라. 
#분석방법론은 CBOW와 Skip-Gram 중 후자를 선택해라.
from gensim.models import Word2Vec
embedding_model_20 = Word2Vec(dic, size=20, window = 2, min_count=2, workers=4, iter=100, sg=1)
```

```python
vocab = list(embedding_model_20.wv.vocab)
vocab
```

```python
print(embedding_model_20['국민'])
```

```python
print(embedding_model_20.wv.most_similar("국민"))
```
